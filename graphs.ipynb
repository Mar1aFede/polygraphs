# Graph views

```python
from polygraphs import graphs
from polygraphs import visualisations as viz
```

```python
g = graphs.complete(6)
viz.draw(g, layout=viz.Layout.CIRCULAR)
```

```python
g = graphs.cycle(6, directed=True)
viz.draw(g, layout=viz.Layout.CIRCULAR)
```

```python
g = graphs.star(6)
viz.draw(g)
```

```python
g = graphs.wheel(6)
viz.draw(g, layout=viz.Layout.SPRING)
```

```python
g = graphs.line(6, directed=True)
viz.draw(g, layout=viz.Layout.CIRCULAR)
```

```python
g = graphs.grid(3, 3)
viz.draw(g)
```

```python
g = graphs.karate()
viz.draw(g)
```

```python
g = graphs.random(10, 0.7, seed=123456789, directed=True)
viz.draw(g)
```
